```python
import random
choose_setting =[
    'forest',
    'library',
    'camping ground',
    'ski lodge',
    '19th century saloon',
    'oasis',
    'carshow'
]
choose_obj = [
    'dog',
    'cat',
    'chicken',
    'ford f150',
          ]
seed = 57
gen_prompt_sequence = {
}
for setting in choose_setting:
    template = f"""A {setting}, filed with interesting realistic objects, environment art,
    landscape art, intricate, insane level of details, artstation HQ"""
    gen_prompt_sequence[template] = random.randrange(20, 5999, 3)
```

```python
outdir ="/content/yolov5/data/"

walk(prompts=list(gen_prompt_sequence.keys()), 
     seeds=list(gen_prompt_sequence.values()),
     make_video=True,
     output_dir = outdir,
     name='images',
     num_steps=100,
     use_lerp_for_text=True,
     disable_tqdm=True,
     resume=False,
     upsample=True
     )
```

```python
def run_ffmpeg(width, height, fps): 
    ffmpg_cmd = [ 'ffmpeg', '-y', '-f', 'rawvideo', '-vcodec', 'rawvideo', '-pix_fmt', 'bgr24', '-s', "{}x{}".format(width, height), '-r', str(fps), '-i', '-', '-hls_time', '5', '-hls_list_size', '6', '{streamResult}/hls/index.m3u8' ] 
    return subprocess.Popen(ffmpg_cmd, stdin=subprocess.PIPE)
```

```python
import ffmpeg
output_options = {
    'crf': 20,
    'preset': 'slower',
    'movflags': 'faststart',
    'pix_fmt': 'yuv420p'
}
(
    ffmpeg
    .input(path, pattern_type='glob', framerate=25)
    .filter_('deflicker', mode='pm', size=10)
    .filter_('scale', size='hd1080', force_original_aspect_ratio='increase')
    .output(
        'movie.mp4', 
        **output_options
    )
    .run()
)
```
